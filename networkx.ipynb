```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()    #graph
# G=nx.DiGraph()      #directed graph
# G=nx.MultiGraph()     #parallel edges between nodes
# G=nx.MultiDiGraph()   #parallel edges but directed
```

```python
G.add_edge(1,2)
G.add_edge(2,3,weight=0.3)
```

```python
nx.draw_spring(G,with_labels=True)
plt.show()
```

```python
edge_list=[(1,2),(2,3),(3,4)]
# G=nx.from_edgelist(edge_list)
G.add_edges_from(edge_list)
nx.draw_spring(G,with_labels=True)
plt.show()
```

```python
print(nx.adjacency_matrix(G))
```

```python
list(G.neighbors(2))
```

```python
import networkx as nx

G = nx.Graph()
G.add_edges_from([
    (1, 2,{'weight':0.2}),
    (2, 3),
    (2, 4)
])
```

```python
nx.draw(G,with_labels=True)
```

```python
list(G.neighbors(2))
```
